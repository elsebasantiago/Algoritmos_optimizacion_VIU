```python
# Actividad Guiada 2, en clase
import numpy as np
import math
import matplotlib.pyplot as plt
import random

def f(x,y):
  return x**2 + y**2

def df(x,y):
  return np.array([2*x, 2*y])


alpha = 0.1
rango = 2.5
resolucion = 100
X = np.linspace(-rango, rango, resolucion)
Y = np.linspace(-rango, rango, resolucion)
Z = np.zeros((resolucion,resolucion))

max_iter = 500
for ix, x in enumerate(X):
  for iy, y in enumerate(Y):
    Z[ix,iy] = f(x,y)

plt.contourf(X,Y,Z,resolucion)
plt.colorbar()

P = [random.uniform(-rango,rango), random.uniform(-rango,rango)]
plt.plot(P[0],P[1],"o", c="white")
print(P)

for _ in range(max_iter):
  grad = df(P[0], P[1])
  P[0], P[1] = P[0] - alpha * grad[0], P[1] - alpha * grad[1]
  plt.plot(P[0],P[1],"o", c="red")

print(f'x = {P[0]}')
print(f'y = {P[1]}')
print(f'grad = {grad} ')
print(f'f = {f(P[0],P[1])} \n')

# Muestra la figura
plt.show()
```

```python
# Actividad Guiada 2, en clase
import numpy as np
import math
import matplotlib.pyplot as plt

def f(P):
  return math.sin((1/2) * P[0]**2 - (1/4) *  P[1]**2 + 3) * math.cos(2 * P[0] + 1 - math.exp(P[1]) )

def df(P):

  h = 0.01
  T = np.copy(P)
  grad = np.zeros(2)
  for it, th in enumerate(P):
    T[it] = T[it] + h
    grad [it] = (f(T) - f(P)) / h

  return grad

alpha = 0.1
rango = 2
max_iter = 1000
x = np.random.uniform(-rango, rango)
y = np.random.uniform(-rango, rango)

punto = np.array([x,y])
print(f'x = {x}')
print(f'y = {y}')
print(f'grad = {df(punto)} ')
print(f'f = {f(punto)} \n')

resolucion = 50
X = np.linspace(-rango, rango, resolucion)
Y = np.linspace(-rango, rango, resolucion)
Z = np.zeros((resolucion,resolucion))

for ix, xx in enumerate(X):
  for iy, yy in enumerate(Y):
    Z[ix,iy] = f(np.array([xx,yy]))

plt.contourf(X,Y,Z,resolucion)
plt.colorbar()

plt.plot( x, y ,"o", c ="white")


for _ in range(max_iter):
  grad = df(np.array([x,y]))
  x = x - alpha * grad[0]
  y = y - alpha * grad[1]
  plt.plot(x,y,"x", c = "red")

plt.show()

print(f'x = {x}')
print(f'y = {y}')
print(f'grad = {grad} ')
print(f'f = {f(np.array([x,y]))} \n')
```
